```python
import pims
import matplotlib.pyplot
import numpy as np
import os
import pandas as pd
import imageio
```

```python
path = 'X:\ibn-vision\DATA\SUBJECTS_AB'
csv_paths = []
directory_list = os.listdir(path)
for directory in directory_list:
    file_list = os.listdir(path + '/' + directory)
    for file in file_list:
        if 'Meta.csv' in file:
            name = pd.read_csv(path + '/' + directory + '/' + file)['file_names']
            csv_paths.append(path + '/' + directory + '/' + file[:-8] + '.avi')
```

```python
Meta_files = [i[:-4]+ 'Meta.csv' for i in csv_paths[:3]]
```

```python
stims_dict = {}
for file in Meta_files:
    session = file[-23:]
    meta = pd.read_csv(file)
    stims = []
    for i in range(1, 7):
        stim_time = meta['stim{}'.format(i)]
        try:
            stims.append(float(stim_time))
        except ValueError:
            stims_dict[session] = stims
            break
        stims_dict[session] = stims
```
